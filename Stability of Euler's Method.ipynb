```python
### MA 341: Project 1, Prompt 1 ###
import numpy as np
import matplotlib.pyplot as plt
import math as math

### Part A ###

#Defining deltat and ts
deltat = 0.1
ts = np.linspace(deltat, 1, 10)
```

```python
#Exact Solution
def decay(k):
    ylist = []
    deltat = 0.1
    for i in range(1, 11):
        yn = math.e**(-k*(deltat*i))
        ylist.append(round(yn, 3))
    return(ylist)
```

```python
#Forward Euler's Method (Approximate Solution)
def euler(k):
    ylist = []
    deltat = 0.1
    tn = 0
    yn = math.e**(-k*tn)
    for i in range(1, 11):
        tn = tn + (deltat*i)
        yn = yn + (-k*yn)*(deltat)
        ylist.append((round(yn, 3)))
    return(ylist)
```

```python
#Plotting Exact Solution with different decay rates
plt.plot(ts, decay(1))
plt.plot(ts, decay(5))
plt.plot(ts, decay(15))
plt.plot(ts, decay(25))
plt.xlabel('Time')
plt.ylabel('Value')
```

```python
#Plotting Forward Euler's Method with different decay rates
plt.plot(ts, euler(1))    
plt.plot(ts, euler(5))
plt.plot(ts, euler(15))
plt.plot(ts, euler(25))
```
