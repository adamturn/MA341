```python
### MA 341: Project 1, Prompt 1 ###
import numpy as np
import matplotlib.pyplot as plt
import math as math

### Part A ###

#Defining deltat and ts
deltat = 0.1
ts = np.linspace(deltat, 1, 10)
```

```python
#Exact Solution
def decay(k):
    ylist = []
    deltat = 0.1
    for i in range(1, 11):
        yn = math.e**(-k*(deltat*i))
        ylist.append(round(yn, 3))
    return(ylist)
```

```python
#Forward Euler's Method (Approximate Solution)
def feuler(k):
    ylist = []
    deltat = 0.1
    tn = 0
    yn = math.e**(-k*tn)
    for i in range(1, 11):
        tn = tn + (deltat*i)
        yn = yn + (-k*yn)*(deltat)
        ylist.append(round(yn, 3))
    return(ylist)
```

```python
#Plotting Exact Solution with different decay rates
plt.plot(ts, decay(1))
plt.plot(ts, decay(5))
plt.plot(ts, decay(15))
plt.plot(ts, decay(25))
plt.xlabel('Time')
plt.ylabel('Value')
```

```python
#Plotting Forward Euler's Method with lower decay rates
plt.plot(ts, feuler(1))    
plt.plot(ts, feuler(5))
```

```python
#Plotting Forward Euler's Method with a higher decay rate
plt.plot(ts, feuler(15))
```

```python
#Plotting Forward Euler's Method with an even higher decay rate
plt.plot(ts, feuler(25))

#NOTE: the behavior of the graphs produced by Forward Euler's
#      method with decay constants '15' and '25' is less than
#      desirable. I do not believe them to be accurate models
#      as they do not show a constant negative slope as one
#      one would expect of a function exhibiting exponential 
#      decay.
```

```python
### Part B ###

#Backward Euler's Method
def beuler(k):
    ylist = []
    deltat = 0.1
    tn = 0
    yn = math.e**(-k*tn)
    for i in range(1,11):
        tn = tn + (deltat*i)
        yn = math.e**(-k*tn) + (-k*(math.e**(-k*tn) + (-k*math.e**(-k*tn)*deltat))*deltat)
        ylist.append(round(yn, 3))
    return(ylist)
```

```python
#Plotting Backward Euler's Method with different decay rates
plt.plot(ts, beuler(1))
plt.plot(ts, beuler(5))    
plt.plot(ts, beuler(15))    
plt.plot(ts, beuler(25))
        
#NOTE: The approximation using Backward Euler appears to have greater
#      stability at higher decay rates than Forward Euler's.
```
